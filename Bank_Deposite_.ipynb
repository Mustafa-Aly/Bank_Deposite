```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline , make_pipeline

from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
from sklearn.preprocessing import OneHotEncoder
import category_encoders as ce

from sklearn.linear_model import LogisticRegression,Lasso,Ridge
from sklearn.svm import SVC
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier


from sklearn.metrics import f1_score,accuracy_score,recall_score,precision_score


from sklearn.decomposition import PCA
from sklearn.feature_selection import SequentialFeatureSelector,SelectFromModel

from sklearn.model_selection import cross_validate,cross_val_score,train_test_split
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

**Encoding**

```python
be_col=['job','month']
ohe_col=['marital','education','default','housing','loan','contact','poutcome']
log_col=['balance']
```

```python
x=df.drop(['deposit','pdays','previous'],axis=1)
y=df['deposit']
```

```python
x
```

```python
y
```

```python
y_dum=pd.get_dummies(y,drop_first=True)
```

```python
y_dum
```

```python
transform=ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(sparse_output=False,drop='first'),ohe_col),
    ('be',ce.BinaryEncoder(),be_col)
],remainder='passthrough')
```

```python
transform
```

```python
x.describe()
```

**Modeling Using Machine Learning**

```python
models=[
    ('lr',LogisticRegression()),
    ('SVC',SVC()),
    ('Knn',KNeighborsClassifier()),
    ('DT',DecisionTreeClassifier()),
    ('Bagging',BaggingClassifier()),
    ('RT',RandomForestClassifier())
]
```

```python
for model in models:
    print(model[0])
    Pipe=make_pipeline(transform,
                      RobustScaler(),
                      model[1])



    cv=cross_validate(estimator=Pipe,X=x,y=y_dum,cv=5,return_train_score=True,scoring='accuracy')
    print(model[0],'train accurry score ',cv['train_score'].mean())
    print(model[0],'test accuracy score  ',cv['test_score'].mean())

    print('-'*30)
```

**Using Random Grid Search**

```python
Pipe=make_pipeline(transform,
                      RobustScaler(),
                      RandomForestClassifier())
param_grid ={
  'randomforestclassifier__bootstrap': [True, False],
 'randomforestclassifier__max_depth': [10, 20, 30, 40, 50, 60, 70, 80],
 'randomforestclassifier__max_features': ['auto', 'sqrt'],
 'randomforestclassifier__min_samples_leaf': [1, 2, 4],
 'randomforestclassifier__min_samples_split': [2, 5, 10],
 'randomforestclassifier__n_estimators': [200, 400, 600, 800, 1000]}
```

```python
gs=RandomizedSearchCV(Pipe,param_grid,n_iter=20,cv=10,return_train_score=True,scoring='accuracy')
```

```python
gs.fit(x , y )
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers.core import Dense,Activation
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
x.shape
```

```python
len(x.columns)
```

```python
def create_model():
  model=keras.Sequential([
      keras.layers.Dense(15,input_shape=(26,),activation='relu'),
      keras.layers.Dense(15,activation='relu'),
      keras.layers.Dense(1,activation='sigmoid'),
  ])
  model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics='accuracy'

                )
  return model
```

```python
model_ann=KerasClassifier(create_model, epochs=200, verbose=0)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y_dum,
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
Pipe=make_pipeline(transform,
                      MinMaxScaler(),
                      model_ann)
```

```python
Pipe.fit(X_train,Y_train)
```

```python
y_pred = Pipe.predict(X_test)
```

```python
y_pred[:10]
```

```python
Y_test[:10]
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,y_pred))
```
